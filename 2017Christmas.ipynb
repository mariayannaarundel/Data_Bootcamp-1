```python
BOROUGH = Christmas_2017['BORO_NM'][Christmas_2017['BORO_NM'] !='UNKNOWN'].dropna()
RACE = Christmas_2017['SUSP_RACE'][Christmas_2017['SUSP_RACE'] !='UNKNOWN'].dropna()
```

```python
Christmas_2017[['BORO_NM','SUSP_RACE','SUSP_AGE_GROUP', 'SUSP_SEX']]
BORO_SUSP=Christmas_2017[['BORO_NM','SUSP_RACE','SUSP_AGE_GROUP', 'SUSP_SEX']].set_index(Christmas_2017[['BORO_NM','SUSP_RACE','SUSP_AGE_GROUP', 'SUSP_SEX']].columns[0])
```

```python
BORO_SUSP
```

```python
BORO_SUSP_C=BORO_SUSP[BORO_SUSP!='UNKNOWN'].dropna()
BORO_SUSP_C
```

```python
QUEENS = BORO_SUSP_C.loc['QUEENS',:]
MANHATTAN = BORO_SUSP_C.loc['MANHATTAN',:]
BRONX = BORO_SUSP_C.loc['BRONX',:]
BROOKLYN = BORO_SUSP_C.loc['BROOKLYN',:]
STATEN = BORO_SUSP_C.loc['STATEN ISLAND',:]
```

```python
QR=QUEENS['SUSP_RACE'].value_counts()
MH = MANHATTAN['SUSP_RACE'].value_counts()
BX = BRONX['SUSP_RACE'].value_counts()
BK = BROOKLYN['SUSP_RACE'].value_counts()
SI = STATEN['SUSP_RACE'].value_counts()
```

```python
fig, (ax1,ax2)  = plt.subplots(1,2, figsize=(18,10))
QR.plot(ax=ax1, kind='pie')
MH.plot(ax=ax2, kind='pie')

ax2.set_title('SUSPECT RACE DISTRIBUTION : QUEENS')

#ax2.set_ylabel('RACE')
```

```python
fig  = plt.figure(figsize=(30,60))
plt.rc('font', size=30)

ax1 = plt.subplot2grid((3,2),(0,0))
QR.plot(ax=ax1, kind='pie')
ax1.set_ylabel('')
ax1.set_title('SUSPECT RACE DISTRIBUTION : QUEENS')

ax1 = plt.subplot2grid((3,2), (0, 1))
MH.plot(ax=ax1, kind='pie')
ax1.set_ylabel('')
ax1.set_title('SUSPECT RACE DISTRIBUTION : Manhattan')

ax1 = plt.subplot2grid((3,2), (2, 0))
SI.plot(ax=ax1, kind='pie')
ax1.set_ylabel('')

ax1 = plt.subplot2grid((3,2), (1, 1))
BX.plot(ax=ax1, kind='pie')
ax1.set_ylabel('')

ax1 = plt.subplot2grid((3,2), (1, 0))
BK.plot(ax=ax1, kind='pie')
ax1.set_ylabel('')
```
